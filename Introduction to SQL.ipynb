# 10.What are the types of join and explain each?

1. **Cross join**

CROSS JOIN returns the Cartesian product of rows from tables in the join. In other words, it will      produce rows which combine each row from the first table with each row from the second table.

```python
SELECT column_name(s)
  
  FROM table1 CROSS JOIN table2;
```

2. **Inner join**
 
An inner join demands that each row in the two connected tables have the same column values, and it is a typical join procedure in applications, but it is not always the best option. Inner join combines column values from two tables (A and B) depending on the join-predicate to create a new result table.

```python
SELECT column_name(s)

 FROM table1

 INNER JOIN table2

ON table1.column_name = table2.column_name;
 
```

3. **Outer join**

The joined table retains each row—even if no other matching row exists. Outer joins subdivide further into left outer joins, right outer joins, and full outer joins, depending on which table's rows are retained: left, right, or both (in this case left and right refer to the two sides of the JOIN keyword).

```python
#left outer join

SELECT column_name(s)

FROM table1

LEFT JOIN table2

ON table1.column_name = table2.column_name;
```

```python
#right outer join

SELECT column_name(s)

FROM table1

RIGHT JOIN table2

ON table1.column_name = table2.column_name;
```

```python
#Full outer join

SELECT column_name(s)

FROM table1

FULL OUTER JOIN table2

ON table1.column_name = table2.column_name

WHERE condition;
```

4. **Self join**
A self-join is joining a table to itself


```python
SELECT column_name(s)
   
 FROM table1 T1, table1 T2

 WHERE condition;
```

# 11.What is normalization?

Normalization is a database design technique that avoids undesired characteristics such as Insertion, Update, and Deletion Anomalies by reducing data redundancy. Normalization rules break huge tables into smaller tables and use relationships to connect them. The goal of SQL normalisation is to reduce duplicate (repetitive) data and ensure logical data storage.

# 12. What is Denormalization?
Denormalization is a database optimization technique in which duplicated data is added to one or more tables. In a relational database, this can help us avoid costly joins. It's important to note that denormalization does not imply the absence of normalization. It is an optimization technique that is used once normalization has been completed.

# 13b.The normal forms can be divided into 5 forms
1. Unnormalized form (UNF) 
Unnormalized form (UNF), also known as an unnormalized relation or non first normal form (N1NF or NF2) in database normalization, is a database data model (data organization in a database) that does not meet any of the database normalization constraints defined by the relational model. Non-relational or NoSQL databases are database systems that support data that is not normalized.


2. Elementary key normal form (EKNF)
Because elementary key normal form (EKNF) is a slight improvement on third normal form, EKNF tables are always in 3NF. When there are multiple distinct compound keys that overlap, this occurs. In such circumstances, the overlapping column may include redundant data (s).


3. Boyce–Codd normal form (BCNF)
The Boyce–Codd normal form (or BCNF or 3.5NF) is a database normalization form. It's a more powerful variation of the third normal form (3NF). Raymond F. Boyce and Edgar F. Codd created BCNF in 1974 to handle some types of anomalies not addressed by 3NF as initially established.

4. Essential tuple normal form (ETNF)
It is an improved version of the third normal form, hence EKNF is in 3rd Normal Form by default. There is redundancy in the overlapping column when there are more than one unique compound key and the keys overlap. Thus, if every non-trivial functional dependency in the 3NF relation requires either a superkey or an elementary key's subkey, it is in EKNF.
5. Domain-key normal form (DKNF)
The domain-key normal form (DK/NF) is a database normalization normal form that requires the database to have no constraints other than domain and key constraints. A domain constraint describes the values that are acceptable for a particular property, whereas a key constraint specifies the attributes that uniquely identify a row in a table.

# Primary Index
 A primary index is an ordered file with two fields and records of a predetermined length. The primary key of the data file is replicated in an ordered way in the first field of the index, and the second field of the ordered file contains a pointer to the data-block where a record containing the key is available.
 
 The primary Index is also further divided into two types.
​
* Dense Index
In a dense index, a record is created for every search key valued in the database. This helps you to search faster but needs more space to store index records. In this Indexing, method records contain search key value and points to the real record on the disk.
​
* Sparse Index
It is an index record that appears for only some of the values in the file. Sparse Index helps you to resolve the issues of dense Indexing in DBMS. In this method of indexing technique, a range of index columns stores the same data block address, and when data needs to be retrieved, the block address will be fetched.
​
# Secondary Index
The secondary Index in DBMS can be generated by a field which has a unique value for each record, and it should be a candidate key. It is also known as a non-clustering index.
​
This two-level database indexing technique is used to reduce the mapping size of the first level. For the first level, a large range of numbers is selected because of this; the mapping size always remains small.
​
#  Clustering Index
In a clustered index, records themselves are stored in the Index and not pointers. Sometimes the Index is created on non-primary key columns which might not be unique for each record. In such a situation, you can group two or more columns to get the unique values and create an index which is called clustered Index. This also helps you to identify the record faster.

# 18. What is a relationship and what are they? 
In the context of databases, a relationship is a situation in which one table has a foreign key that references the primary key of the other table. Relationships enable relational databases to partition and store data in various tables while also connecting distinct data pieces.
# Types of relationships
*** One-to-One Relationship**

Such a relationship exists when each record of one table is related to only one record of the other table.

*** One-to-Many or Many-to-One Relationship**

Such a relationship exists when each record of one table can be related to one or more than one record of the other table. This relationship is the most common relationship found. A one-to-many relationship can also be said as a many-to-one relationship depending upon the way we view it.

*** Many-to-Many Relationship**

Such a relationship exists when each record of the first table can be related to one or more than one record of the second table and a single record of the second table can be related to one or more than one record of the first table. A many-to-many relationship can be seen as a two one-to-many relationship which is linked by a 'linking table' or 'associate table'. The linking table links two tables by having fields which are the primary key of the other two tables.

# 24 What is the difference between DELETE and TRUNCATE commands?
**DELETE** is a DML (Data Manipulation Language) command that is used to specify which row(tuple) from a table or relation we wish to remove or delete. A WHERE clause can be included in the DELETE command. When the WHERE clause is used with the DELETE command, only those rows(tuples) that satisfy the condition are removed or deleted; otherwise, all tuples(rows) from the table are removed by default.

```python
DELETE FROM employees WHERE condition;
```


**TRUNCATE** is a Data Definition Language (DDL) command that deletes all of the rows or tuples in a table. The TRUNCATE command, unlike the DELETE command, does not have a WHERE clause. The transaction log for each deleted data page is recorded by the TRUNCATE command. The TRUNCATE command is faster than the DELETE command. Even after using the TRUNCATE command, we can rollback the data, just like we can with DELETE.


```python
TRUNCATE TABLE  employees;
```
